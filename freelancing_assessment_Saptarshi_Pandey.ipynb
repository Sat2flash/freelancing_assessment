```python
# Import necessary libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
import os
```

```python
# Load the dataset from a CSV file
df = pd.read_csv("C:/Users/sapta/Downloads/device_features.csv")
```

```python
df
```

```python
# Set up the API key and parameters for the currency conversion API
# For this assessment openexchangerates api has been used 
%load_ext dotenv
%dotenv

api_key = os.getenv('OPENEXCHANGERATES_API_KEY')
exchange_params = {
    "app_id" : api_key,
    "base" : "USD"
}
print("API Key:", api_key)
```

```python
# Fetch the latest exchange rates
response = requests.get(url= "https://openexchangerates.org/api/latest.json", params =exchange_params)
data = response.json()
rates_dictionary = data["rates"]
rates_dictionary
```

```python
# Function to convert prices to USD
def convert_to_usd(row):
    currency = row['price_currency']
    price = row['price']
    # Dividing by the exchange rate converts the price to USD
    return price / rates_dictionary.get(currency, 1)
```

```python
# Apply the function to each row to create a new column with prices in USD
df['price_in_usd'] = df.apply(convert_to_usd, axis=1)
df
```

```python
# List of brands to analyze
brands_list = ['Samsung', 'Motorola', 'Xiaomi', 'Nokia', 'Sony', 'Asus', 'Lenovo', 'T-Mobile', 
               'BBK', 'Apple', 'ZTE', 'Microsoft', 'LG', 'Sharp']
average_prices_usd = {}

# Calculate the average price in USD for each brand
for brand in brands_list:
    brand_df = df[df['brand'] == brand]
    average_prices_usd[brand] = brand_df['price_in_usd'].mean()

# Print the average prices for each brand    
for brand, avg_price in average_prices_usd.items():
    print(f"The average price for {brand} in USD is: {avg_price:.2f}")
```

```python
# Second challenge
# Exchange rate for GBP
exchange_rate_gbp = rates_dictionary["GBP"]
exchange_rate_gbp

# Convert 'released_date' to datetime format and extract year and month
df['released_date'] = pd.to_datetime(df['released_date'], format='%d-%m-%y', errors='coerce')
df['year'] = df['released_date'].dt.year
df['month'] = df['released_date'].dt.month
```

```python
# Function to convert prices to GBP
def convert_to_gbp(row):
    usd_price = row['price_in_usd']
     # Multiplying by the GBP exchange rate converts the price to GBP
    return (usd_price * exchange_rate_gbp)

# Apply the function to each row to create a new column with prices in GBP
df['price_in_gbp'] = df.apply(convert_to_gbp, axis=1)
df
```

```python
# Filtering data for years 2020 to 2023 and calculating average monthly prices
years = [2020, 2021, 2022, 2023]
monthly_avg_prices = {}

for year in years:
    year_df = df[df['year'] == year]
    monthly_avg = year_df.groupby('month')['price_in_gbp'].mean()
    monthly_avg_prices[year] = monthly_avg
```

```python
# Plotting the charts for each year
for year, avg_prices in monthly_avg_prices.items():
    plt.figure(figsize=(10, 6))
    avg_prices.plot(kind='bar', color='skyblue')
    plt.title(f'Monthly Average Price Trends in GBP for {year}')
    plt.xlabel('Month')
    plt.ylabel('Average Price (GBP)')
    plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.show()
```
